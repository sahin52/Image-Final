```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt



```

```python
def accentuateTheRoads(img):
    return cv.Canny(cv.cvtColor(img, cv.COLOR_RGB2GRAY),50,150,apertureSize=3)
```

```python
img = cv.imread("Dataset3/1.png")
roads = accentuateTheRoads(img)
plt.imshow(cv.Canny(cv.cvtColor(img, cv.COLOR_RGB2GRAY),50,150,apertureSize=3)
```
