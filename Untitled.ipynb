```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np


img =cv.imread("Dataset2/5.jpg")
#img = plt.imread("Dataset2/5.jpg")
img = cv.cvtColor(np.array(img), cv.COLOR_RGB2BGR)
plt.imshow(img)
plt.show()
```

```python
def removeNonBlackNWhite(img):
    for i in range(0,len(img)):
        for j in range(0,len(img[0])):
            pixel = img[i][j]
            smallest = np.amin(pixel)
            pixel = pixel-smallest
            dif = np.amax(pixel)
            if(dif>45):
                img[i][j]=np.zeros((3))
            if(pixel[0]<pixel[1] - 5 and pixel[2]<pixel[1]-5):
                img[i][j] = np.zeros((3))
    return img
```

```python
img =cv.imread("Dataset2/5.jpg")
#img = plt.imread("Dataset2/5.jpg")
img = cv.cvtColor(np.array(img), cv.COLOR_RGB2BGR)
img = removeNonBlackNWhite(img)
```

```python
plt.imshow(img)
plt.show()
```
